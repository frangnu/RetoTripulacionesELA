```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_sexo = pd.DataFrame({'Sexo':['Hombre','Mujer'],
                        '% donantes por sexo': [69.39,30.61],
                        'Media donanciones en €': [206.63,189.33]                                             
                  })
df_sexo
plt.bar(df_sexo['Sexo'], df_sexo['% donantes por sexo'])
plt.title('Donaciones % por sexo');
```

```python
plt.bar(df_sexo['Sexo'], df_sexo['Media donanciones en €'])
plt.title('Donaciones media en €');
```

```python
df_estado_civil= pd.DataFrame({ 'Estado civil': ['Soltero','Casado','Divorciado','Otros'],
                               '% donantes estado civil':[76.18,12.78,6.13,4.91],
                             'Donación media en €': [196.28,np.NaN,np.NaN,np.NaN]}) 
df_estado_civil
plt.bar(df_estado_civil['Estado civil'], df_estado_civil['% donantes estado civil'])
plt.title('% donación según estado civil');
```

```python
df_vivienda = pd.DataFrame({ 'Vivienda en propiedad': ['Si','No'],
                            '% donantes vivienda en propiedad':[88.51,11.49],
                            })
df_vivienda
plt.bar(df_vivienda['Vivienda en propiedad'], df_vivienda['% donantes vivienda en propiedad'])
plt.title('% donación según propiedad de vivienda');
```

```python
df_descendencia= pd.DataFrame({ 'Descendencia': ['Sin descendencia','2 descendientes','1 descendiente','Otros'],
                               '% donantes Descendencia':[46.19,25.77,19.25,8.79]}) 
df_descendencia

plt.bar(df_descendencia['Descendencia'], df_descendencia['% donantes Descendencia'])
plt.title('% donación según descendencia')
plt.tight_layout();
```

```python
df_actividad = pd.DataFrame({ 'Tipo de actividad económica': ['Cuenta propia','Act. Industriales Ajena','Act. Primaria', 'Otro'],
                               '% donantes Actividad':[73.10,22.00,3.60,1.3]}) 
df_actividad

plt.bar(df_actividad['Tipo de actividad económica'], df_actividad['% donantes Actividad'])
plt.title('% donación según actividad económica')
plt.tight_layout();
```
